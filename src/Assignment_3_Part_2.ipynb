```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("titanic.csv")
df.head()
```

```python
#checking the analysis for the data
#missing values
df.isnull().sum()
```

```python
#Hypothesis 1
#Determine if the survival rate is associated to the class of passenger
titanic = pd.read_csv('titanic.csv')

# group the data by passenger class and calculate the survival rate
class_survival = titanic.groupby('Pclass')['Survived'].mean()

plt.bar(class_survival.index, class_survival.values)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Hypothesis 2
#Determine if the survival rate is associated to the gender
gender_survival = titanic.groupby('Sex')['Survived'].mean()

# plot the survival rate by gender
plt.bar(gender_survival.index, gender_survival.values)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Hypothesis 3
#Determine the survival rate is associated with the age.

titanic = pd.read_csv('titanic.csv')

# group the data by age and calculate the survival rate
age_survival = titanic.groupby('Age')['Survived'].mean()

# plot the survival rate by age
plt.plot(age_survival.index, age_survival.values)
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Survival Rate')
plt.show()
```
